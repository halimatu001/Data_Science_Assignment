```python

file_path = "/content/drive/My Drive/Business Funding Data.csv"

# Try different encodings
df = pd.read_csv(file_path, encoding="ISO-8859-1")  # OR encoding="latin1"

df.head(10)  # Display first few rows

```

### Missing Value

```python
missing_percentage = df.isnull().sum()/len(df)*100
missing_percentage
```

### Drop Financial Type Normalized Column

```python
df.drop(columns=['Financing Type Normalized'], inplace=True)
```

### Drop Amount

```python
df.drop(columns=['Amount'], inplace=True)
```

### Drop Investor Count


```python
df.drop(columns=['Investors Count'], inplace=True)
```

### Check for info

```python
df.info()
```

### Fill in missing value for effective date with mode

```python
df['Effective date'].fillna(df['Effective date'].mode()[0], inplace=True)
```

### Fill in missing value for financing type with mode

```python
df['Financing Type'].fillna(df['Financing Type'].mode()[0], inplace=True)
```

### Fill in missing value for investor with Unknown

```python
df['Investors'].fillna("Unknown", inplace=True)
```

### Handling duplicate

```python
duplicate = df.duplicated().sum()
print(f"Number of duplicated rows: {duplicate}")
```

### Removing Outliers

```python
df.head(10)
```

```python
mean = df["Amount Normalized"].mean()
std_dev = df["Amount Normalized"].std()

# Calculate Z-score
Z_Score = np.abs((df["Amount Normalized"] - mean) / std_dev)

# Filter outliers (Z-score < 3)
df_no_outliers = df[Z_Score < 3]

# Display the first few rows
df_no_outliers.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
categorical_cols = ['Website Domain', 'Financing Type', 'Categories', 'Investors']

for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col].astype(str))
```

```python
display(df.head())
```

```python
#Save cleaned dataset
df.to_csv('/content/drive/My Drive/Cleaned_Business_Funding_Data.csv', index=False)
```

### Observations from Exploring the Data
- The dataset contained missing values in 'Effective date', 'Financing Type', 'Investors', and 'Amount Normalized'.
- Some columns ('Financing Type Normalized' and 'Amount') were redundant and removed.
- Duplicates were identified and removed.
- Outliers in 'Amount Normalized' were removed using the Z-score method.

### Steps Taken for Cleaning and Preprocessing
1. **Handled Missing Values:**
   - Used mode for categorical variables (e.g., 'Effective date').
   - Used median for numerical variables ('Amount Normalized').
   
2. **Dropped Redundant Columns:**
   - Removed unnecessary columns that duplicated information.
   
3. **Removed Duplicates:**
   - Ensured unique rows for more accurate analysis.
   
4. **Handled Outliers:**
   - Applied Z-score to remove extreme values.
   
5. **Encoded Categorical Variables:**
   - Used Label Encoding for categorical features.

### Justifications for Each Step
- Mode was used for categorical missing values to maintain consistency.
- Median was chosen for 'Amount Normalized' because it is less affected by extreme values.
- Label Encoding was used for categorical columns to prepare for machine learning models.
- Dropping duplicate and redundant columns made the dataset cleaner and more efficient for analysis.

### Reflections on Data Preprocessing
Data preprocessing is crucial in real-world analysis because raw datasets often contain inconsistencies, missing values, and outliers. Proper cleaning and transformation ensure that analysis and machine learning models provide meaningful insights.
